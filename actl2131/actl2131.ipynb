# Illustrating CLT

```python
import numpy as np
from numpy.random import gamma
import matplotlib.pyplot as plt
```

```python
def generate_sample_mean(alpha, beta, n):
    sample = gamma(alpha, 1/beta, n)

    sample_mean = np.sum(sample) / n

    return sample_mean
```

```python
alpha = 1
beta = 1
n = 10_000

sample_means = [generate_sample_mean(alpha, beta, n) for _ in range(1_000)]
plt.hist(sample_means, bins=100)
plt.xlim([0.95, 1.05])
plt.title('1,000 Sample Means of size n=10,000 of X ~ Gamma(1,1)')

plt.show()
```

```python
n = 1_000_000
sample_means = [generate_sample_mean(alpha, beta, n) for _ in range(1_000)]
plt.hist(sample_means, bins=100)
plt.xlim([0.95, 1.05])
plt.title('1,000 Sample Means of size n=1,000,000 of X ~ Gamma(1,1)')

plt.show()
```

```python
alpha = 0.4
beta = 0.4
n = 10_000

sample_means = [generate_sample_mean(alpha, beta, n) for _ in range(1_000)]
plt.hist(sample_means, bins=100)
plt.xlim([0.95, 1.05])
plt.title('10,000 Sample Means of size n=10,000 of Y ~ Gamma(0.4, 0.4)')

plt.show()
```
